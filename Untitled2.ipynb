#### .describe() method provides a statistical summary of the data in your DataFrame. It provides the following information for numerical columns:

Standard Deviation: the measure of the spread of data.
Minimum: the minimum value.

Median: the middle value, separating the higher and lower halves of the data.

## EDA - Exploratory Data Analysis

```python
sns.countplot(x='Pclass', data=titanic)
titanic['Pclass'].mode()
```

Conclusion of above plot: more people were there in class 3 as compared to class 1 and class 2

```python
sns.countplot(x='Survived', data=titanic)
```

Conclusion of above plot: More people died i.e. 0 and less people survived i.e. 1 in this ship.

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic)
```

Conclusion of above plot: Most of the people in class 1. And majority of the people died were from class 3.

```python
sns.countplot(x='Sex', hue='Survived', data=titanic)
```

Conclusion of above plot: The no of Males were more than female. But more female survived.

```python
sns.countplot(x='Embarked',hue='Survived',data=titanic)
```

Conclusion of above plot: People who embarked from S and Q were most likely to not survive and People who embarked from C had more chance of survival.

```python
sns.violinplot(x='Pclass',y='Sex',data=titanic)
```

Conclusion of above plot: There were more Female in Pclass 1 than Male in Pclass 1. Also there were more Male in Pclass 3 than Female in same class.

```python
sns.violinplot(x='Pclass', data=titanic)
```

Conclusion of above plot: Maximum people were in Pclass 3 followed by Pclass 1 and then least in Pclass 2.

```python
sns.violinplot(x='Pclass',y='Sex',hue='Survived',data=titanic)
```

Conclusion of above plot: Male and Female in Pclass 3 did not survive the most.

```python
sns.barplot(x='Pclass',y='Fare',data=titanic)
```

Conclusion of above plot: Pclass 1 had the highest fare then Pclass 2 followed by Pclass 3.

```python
sns.stripplot(x='Embarked', y='Fare', data=titanic)
```

Conclusion of above plot: Maximum fare was of people who embarked from C then from S followed by Q.
