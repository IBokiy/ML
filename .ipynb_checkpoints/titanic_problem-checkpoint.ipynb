The columns having null values are: Age, Cabin, Embarked. They need to be filled up with appropriate values later on.

Features: The titanic dataset has roughly the following types of features:

Categorical/Nominal: Variables that can be divided into multiple categories but having no order or priority. 
Eg. Embarked (C = Cherbourg; Q = Queenstown; S = Southampton)
Binary: A subtype of categorical features, where the variable has only two categories. 
Eg: Sex (Male/Female)
Ordinal: They are similar to categorical features but they have an order(i.e can be sorted). 
Eg. Pclass (1, 2, 3)
Continuous: They can take up any value between the minimum and maximum values in a column. 
Eg. Age, Fare
Count: They represent the count of a variable. 
Eg. SibSp, Parch
Useless: They don’t contribute to the final outcome of an ML model. Here, PassengerId, Name, Cabin and Ticket might fall into this category.

## Missing value treatment

```python
titanic.isnull().sum()
```

```python
titanic.groupby('SibSp')['Age'].describe()
```

```python
titanic.groupby('SibSp')['Age'].median()
```

```python
titanic['Age'].median()
```

```python
titanic['Age']=np.where((titanic['SibSp']==8) & (titanic['Age'].isnull()),28.0,titanic['Age'])
titanic['Age']=np.where((titanic['SibSp']==0) & (titanic['Age'].isnull()),29.0,titanic['Age'])
titanic['Age']=np.where((titanic['SibSp']==1) & (titanic['Age'].isnull()),30.0,titanic['Age'])
titanic['Age']=np.where((titanic['SibSp']==2) & (titanic['Age'].isnull()),23.0,titanic['Age'])
titanic['Age']=np.where((titanic['SibSp']==3) & (titanic['Age'].isnull()),10.0,titanic['Age'])
```

```python
titanic.isnull().sum()
```

All missing values in the 'Age' column have been replaced.

## Dividing the numerical and categorical variables 

```python
df_num = titanic.select_dtypes(include=[np.number]).copy()
```

```python
df_num
```

## Dummy variable for categorical data

```python
df_cat = titanic.select_dtypes(include='object').copy()
```

```python
df_cat
```

```python
df_cat = pd.get_dummies(df_cat)
```

```python
df_cat
```

## Combining the numrical and categorical dataframes

```python
df = pd.concat([df_num,df_cat],axis=1)
```

```python
df
```

```python
df.drop(columns=['Sex_male','Embarked_C','PassengerId'],inplace=True)
```

```python
df
```

```python
print(df.info())
print(df.head())
```

## ML Algorithms Start

Create 2 new date frames and assign them to variables:<br>
X - contains the input features;<br>
Y - contains the target set.<br>
In general, the code divides the data into two variables: X, which contains all columns except 'Survived', and Y, which contains only the 'Survived' column. 

```python
X = df.drop(columns=['Survived_0', 'Survived_1'])
Y = df['Survived_1']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=.30,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
Y_pred = lr.predict(X_test)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,Y_pred)
print('Accuracy is',"{:.2f}%".format(100*accuracy))
```
